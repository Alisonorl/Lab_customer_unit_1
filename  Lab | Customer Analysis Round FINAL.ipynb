```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

import warnings
warnings.filterwarnings('ignore') # no more 'useless' red warnings 
```

```python
#01 - Problem (case study)

#Make the marketing campain the most efficient to have more donation based on data
```

```python
#02 - Getting Data

data = pd.read_csv('marketing_customer_analysis.csv')
data.head()
```

```python
#03 - Cleaning/Wrangling/EDA
#Change headers names.
#Deal with NaN values.
#Categorical Features.
#Numerical Features.
#Exploration.

data.columns = data.columns.str.lower()
data[data.isna()]
```

```python
data=data.fillna(0)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
numerical = data.select_dtypes(include=np.number)
numerical
```

```python
categorical = data.select_dtypes(exclude=np.number)
categorical
```

```python
#04 - Processing Data
#Dealing with outliers.
#Normalization.
#Encoding Categorical Data.
#Splitting into train set and test set.
```

```python
y = data['total claim amount']
X = data.drop(['total claim amount'], axis=1)
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = 'object')
cat_data = pd.get_dummies(X, drop_first=True)
cat_data
```

```python
encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(X_cat)
```

```python
encoded = encoder.transform(X_cat).toarray()
encoded
```

```python
encoder.categories_
```

```python
x = np.concatenate([X_num, encoded], axis=1)
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#05 - Modeling
#Apply model.
```

```python
prediction = model.predict(x_test)
prediction.shape
```

```python
#06 - Model Validation
#R2.
#MSE.
#RMSE.
#MAE.
```

```python
r2_score(y_test, prediction), mean_squared_error(y_test, prediction, squared=False), mean_squared_error(y_test, prediction)
```
